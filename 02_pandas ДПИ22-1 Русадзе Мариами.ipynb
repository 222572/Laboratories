```python
# 1.1
import pandas as pd
import numpy as np
from datetime import datetime
recipes = pd.read_csv(path + "recipes_sample.csv", sep=',')
reviews = pd.read_csv(path + "reviews_sample.csv", sep=',')
reviews.rename(columns = {"Unnamed: 0": "index"}, inplace=True)
reviews = reviews.set_index("index")
```

```python
# 1.2
print("Таблица recipes:")
print(len(recipes))
print(len(recipes.columns))
print(recipes.dtypes, '\n')

print("Таблица reviews:")
print(len(reviews))
print(len(reviews.columns))
print(reviews.dtypes)
```

```python
# 1.3
print("Таблица recipes:")
print(recipes.isnull().sum())
print("Доля пропусков:", sum([True for idx, row in recipes.iterrows() if any(row.isnull())]) / len(recipes), '\n')

print("Таблица reviews:")
print(reviews.isnull().sum())
print("Доля пропусков:", sum([True for idx, row in reviews.iterrows() if any(row.isnull())]) / len(reviews))
```

```python
# 1.4
print("Таблица recipes:")
print(recipes[["id", "minutes", "contributor_id", "n_steps", "n_ingredients"]].mean(), "\n")

print("Таблица reviews:")
print(reviews[["user_id", "recipe_id", "rating"]].mean())
```

```python
# 3.1
recipes["description_length"] = recipes["description"].str.len()
recipes
```

```python
# 3.2
recipes["name"] = recipes["name"].str.title()
recipes
```

```python
# 3.3
recipes["name_word_count"] = recipes["name"].str.split().str.len()
recipes
```

```python
# 4.1
group_contributor = recipes.groupby("contributor_id", sort=False)
print(group_contributor.size())
print(group_contributor.size().idxmax())
```

```python
# 4.2
group_id = reviews.groupby("recipe_id", sort=False)
print(group_id["rating"].mean())
print(sum(group_id.apply(lambda x: x["review"].isnull().sum())))
```

```python
# 4.3
print(recipes.groupby(pd.Grouper(key="submitted", freq="Y")).size())
```

```python
# 5.1
merge_df = recipes.merge(reviews, left_on="id", right_on="recipe_id", how="inner")
merge_df = merge_df[["id", "name", "user_id", "rating"]]
merge_df
```

```python
# 5.2
merge_df_2 = recipes.merge(reviews, left_on="id", right_on="recipe_id", how="left")
# nan_reviews = merge_df_2[merge_df_2["review"].isnull()]["id"]
merge_df_2 = merge_df_2[["recipe_id", "name"]]
merge_df_2 = merge_df_2.groupby(["recipe_id", "name"]).size().to_frame(name = 'review_count')
merge_df_2
```

```python
# 5.3
merge_df_3 = recipes.merge(reviews, left_on="id", right_on="recipe_id", how="inner")
merge_df_3 = merge_df_3.groupby(pd.Grouper(key="submitted", freq="Y"))["rating"].mean()
merge_df_3.loc[merge_df_3 == merge_df_3.min()].index[0].year
```

```python
# 6.1
recipes = recipes.sort_values(by="name_word_count", ascending=False)
recipes.to_csv(path + "name_word_count.csv")
```
